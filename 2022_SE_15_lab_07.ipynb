# Normalize X_test to center around 0

```python

X_test_normalized = (X_test - np.mean(X_test)) / np.std(X_test)
```

# Sigmoid function

```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

# Apply sigmoid function to normalized values

```python
Y_pred = sigmoid(X_test_normalized)
```

# Plot the correct sigmoid curve
